```python
import numpy as np
import matplotlib.pyplot as plt
from skopt.sampler import Lhs

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

def plot_lhs_sampling():
    # Define the bounds for the two variables
    bounds = [(0.001, 0.010), (-0.01255805, 0.0)]
    
    # Create the Lhs sampler object
    lhs = Lhs(lhs_type="classic", criterion="maximin")
    
    # Generate the Latin hypercube sampling of 30 samples
    samples = lhs.generate(dimensions=bounds, n_samples=30, random_state=42)
    samples = np.array(samples)
    # Extract the two variables t and C from the samples
    t = samples[:, 0]
    C = samples[:, 1]
    
    # Plot the samples in a scatter plot
    fig, ax = plt.subplots(figsize=(3,3))
    ax.scatter(t, C, c="blue", alpha=0.5)
    ax.set_xlabel(r"$t_w$")
    ax.set_ylabel(r"$CP3_y$")
    #ax.set_title("Latin Hypercube Sampling of 30 Samples")
    plt.tight_layout()
    plt.savefig('./figures/lhs_sampling.pdf')
    plt.show()
```

```python
plot_lhs_sampling()
```
