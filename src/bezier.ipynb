```python
from bezier import bezier_curve
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

cps = np.loadtxt('baselineCP.txt', delimiter=',')
```

```python
fig, ax = plt.subplots(figsize=(3,3))
t_w = 0.005
ax.plot(bezier_curve(cps)[:,0], bezier_curve(cps)[:,1], color='k')
ax.plot(bezier_curve(cps)[:,0], bezier_curve(cps)[:,1]+t_w, color='k')
ax.plot(bezier_curve(cps)[:,0], -bezier_curve(cps)[:,1], color='k')
ax.plot(bezier_curve(cps)[:,0],0-bezier_curve(cps)[:,1]-t_w, color='k')
ax.plot(cps[:,0], cps[:,1], marker='o', ls='--', c="blue", alpha=0.5)
#ax.annotate(r'CP3_y', (cps[2,0], cps[2,1]), )
ax.text(cps[2,0]+0.005, cps[2,1], r'$CP3_y$')

ax.set_ylabel(r'$y\; [m]$')
ax.set_xlabel(r'$x\; [m]$')
ax.axis('equal')
plt.tight_layout()
plt.savefig('./figures/nozzle_shape.pdf')
plt.show()
```
